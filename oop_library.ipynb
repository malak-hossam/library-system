```python
import pandas as pd
import csv
import os

class book():
  def __init__(self, title, author, genre,publication_year):
    self.title = title
    self.author = author
    self.genre = genre
    self.publication_year = publication_year

class library():
  def __init__(self):
    self.dic = {}
    self.count = 0
    self.load_from_file()
    self.display_all_books()

  def add(self, book):
    if book.title not in [details['title']for _,details in self.dic.items()]:
      self.count += 1
      self.dic[self.count] = {'title':book.title, 'author': book.author, 'genre': book.genre, 'year': book.publication_year}
    else:
      print("the boook is already here")

  def remove(self,title):
    for key,details in list(self.dic.items()):
      if details['title'] == title:
        del self.dic[key]
        print(f"the book {title} has been removed")
        break
    else:
      print("the book isn't in the library")

  def search(self, item):
      found_book = {key: details for key, details in self.dic.items() if item in details['title'] or item in details['author']}
      if found_book:
          for key, details in found_book.items():
              print(f"Books found: {details['title']}: Author: {details['author']}, Genre: {details['genre']}, Year: {details['year']}")
      else:
          print("No books found with that title or author.")

  def display(self):
    for key, details in self.dic.items():
      print(f"{key}: {details}")

  def open_file(self):
    with open("library.csv", 'a') as f:
      for key, details in self.dic.items():
        writer = csv.writer(f)
        writer.writerow(list(details.values()))

  def write_header(self):
    with open("library.csv", 'w') as f:
      header = ['title', 'author', 'genre', 'year']
      writer = csv.writer(f)
      writer.writerow(header)

  def load_from_file(self):
    if os.path.exists("library.csv"):
      data = pd.read_csv('library.csv')
      for _, row in data.iterrows():
        self.count += 1
        self.dic[self.count] = {'title': row['title'],'author': row['author'],'genre': row['genre'],'year': row['year']}
      else:
        self.write_header()

  def display_all_books(self):
    print("Existing books in the library:")
    for key, details in self.dic.items():
      print(f"{key}: Title: {details['title']}, Author: {details['author']}, Genre: {details['genre']}, Year: {details['year']}")


def value_error(str):
  while True:
    try:
      return int(input(str))
    except:
      print("Error: you need to enter integer numbers only,please try again\n")

def operation():
  b = input("please enter book name\n")
  a = input("please enter author name\n")
  g = input("please enter genre name\n")
  y = value_error("please enter the year\n")
  return book(b,a,g,y)

s = library()
state = True
while state:
  oper = value_error("availabe operations:\n1: add book\n2: remove book\n3: search for a book\n4: displaythe library\nenter the number of the operation you want\n")
  if oper == 1:
    b = operation()
    s.add(b)
  elif oper == 2:
    title = input("Please enter the book name to remove: ")
    s.remove(title)
  elif oper == 3:
    item = input("please enter the title or the author for the book you search for: ")
    s.search(item)
  elif oper == 4:
    s.display()
  else:
    print("you entered invalid number, please try again")
    continue

  ask = value_error("1: start new operation\n2: End\nenter the number of the option you want\n")
  if ask != 1:
    state = False
  else:
    continue
s.open_file()
```
